```python
#=======================================================
# EXPENSE TRACKER - CLI version (command line interface)
#  Built by: Reshika
#  What it does: Track your daily spends, set budgets per
#  category, get warnings before you go broke!
# ======================================================

import pandas as pd                 # for reading and writing data to csv files
import json                         # for reading/writing budget configurations
import os                           # for checking if files exist
from datetime import date           # for getting today's date
import matplotlib.pyplot as plt     # for visualizing spending patterns

# -------------------------------------------------------
# FILE PATHS
# All data is stored in these two files
# -------------------------------------------------------
EXPENSES_FILE="expenses.csv"     # stores every expense you add
BUDGET_FILE="budget.json"        # stores your monthly budgets

# -------------------------------------------------------
# CATEGORIES
# These are the spending categories you can choose from
# You can add more categories as needed
# -------------------------------------------------------
CATEGORIES=[
    "Food",
    "Travel",
    "Shopping",
    "Entertainment",
    "Education",
    "Health",
    "miscellaneous"
]

# ============================================================
# SECTION 1 — FILE SETUP
# These functions make sure data files exist before we use them
# ============================================================
def setup_files():
     """
    Runs once when the app starts.
    Creates expenses.csv and budget.json if they don't exist yet.
    """
    #----Create expenses.csv if it doesn't exist----
    if not os.path.exists(EXPENSES_FILE):
        #make an empty DataFrame with these 4 columns
        df=pd.DataFrame(columns=["Date","Amount","Category","Note"])
        df.to_csv(EXPENSES_FILE,index=False) # save it to a csv file
        print("Created expenses.csv to store your expenses.")

    #----Create budget.json if it doesn't exist----
    if not os.path.exists(BUDGET_FILE):
        #Default budget: Rs 0 for every category (user will set their own)
        default_budget={category:0 for category in CATEGORIES}
        with open(BUDGET_FILE,"w") as f:
            json.dump(default_budget,f,indent=4)  # save the default budget to a json file
        print("Created budget.json to store your monthly budgets.")

# ============================================================
# SECTION 2 — BUDGET FUNCTIONS
# Set and read your monthly budget per category
# ============================================================

def load_budget():
    """
    Reads budget.json and returns it as a Python dictionary.
    Example: {"Food": 2000, "Travel": 500, ...}
    """
    with open(BUDGET_FILE,"r") as f:
        return json.load(f)

def save_budget(budget):
    """
    Saves the budget dictionary back into budget.json.
    Called after the user updates any budget value.
    """
    with open(BUDGET_FILE,"w") as f:
        json.dump(budget,f,indent=4)

def set_budget():
    """
    Lets the user set a monthly budget for each category.
    Loops through every category and asks for an amount.
    """
    print("\n---Set Your Monthly Budgets---")
    print("Enter Rs amount for each category(press Enter to skip):\n")

    budget=load_budget() # load exsisting budget values

    for category in CATEGORIES:
        current= budget[category]  #what's already saved
        user_input=input(f"{category} (current: Rs{current}):").strip()

        if user_input=="":
            #user pressed Enter - keep the old value 
            continue
        elif user_input.isdigit():
            #user typer a number- update the budget
            budget[category]=int(user_input)
        else:
            print(f" Invalid input for {category}. Keeping old value.")
            save_budget(budget)  # save the updated budget back to the file
            print("\nBudget saved successfully!")

# ============================================================
# SECTION 3 — EXPENSE FUNCTIONS
# Add, view, and analyse your expenses
# ============================================================
def load_expenses():
    """
    Reads expenses.csv and returns a Pandas DataFrame.
    If the file is empty, returns an empty DataFrame.
    """
    df= pd.read_csv(EXPENSES_FILE)

    # Convert the Date column from string to actual date type
    # This lets us filter by month later
    if not df.empty:
        df["Date"]=pd.to_datetime(df["Date"])

    return df

def add_expenses():
    """
    Asks the user to enter one expense and saves it to CSV.
    Fields: Date (auto), Amount, Category, Note
    """
    print("\n---Add a New Expense---")

    #----Amount----
    while True:
        amount_input=input(" Amount(Rs): ").strip()
        if amount_input.isdigit() and int(amount_input)>0:
            amount=int(amount_input)
            break
        else:
            print("please enter a valid positive number.")

    #----Category----
    print("\n Categories:")
    for i, cat in enumerate(CATEGORIES,start=1):
        print(f"   {i}.  {cat}")

    while True:
        cat_input=input("\n Choose a category number:").strip()
        if cat_input.isdigit() and 1 <= int(cat_input) <=len(CATEGORIES):
            category= CATEGORIES[int(cat_input)-1]
            break
        else:
            print(f" enter a number between 1 and {len(CATEGORIES)}.")

    #--- Note ---
    note= input("short note(e.g. Swiggy dinner):").strip()
    if note=="":
        note="No note"

    #--- Date(auto = today) ---
    today=date.today().strftime("%Y-%m-%d") #format:2025-05-03
    


            























```
