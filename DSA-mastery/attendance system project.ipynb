```python
from datetime import datetime
import os

# File names for storing data
STUDENT_FILE = "students.txt"
ATTENDANCE_FILE = "attendance.txt"

def load_students():
    """Loads students from the text file into a dictionary."""
    students = {}
    if not os.path.exists(STUDENT_FILE):
        return students  # Return empty dictionary if file doesn't exist yet
        
    with open(STUDENT_FILE, "r") as file:
        for line in file:
            line = line.strip()  # Remove newline character (\n)
            if line:
                # Split the line by the comma
                student_id, name = line.split(",")
                students[student_id] = name
    return students

def add_student():
    """Registers a new student by appending to the text file."""
    students = load_students()
    print("\n--- Register New Student ---")
    student_id = input("Enter unique Student ID: ").strip()
    
    if student_id in students:
        print("❌ Error: A student with this ID already exists!")
        return
        
    name = input("Enter Student Name: ").strip()
    
    if student_id == "" or name == "":
        print("❌ Error: ID and Name cannot be empty.")
        return

    # 'a' mode appends data to the end of the text file
    with open(STUDENT_FILE, "a") as file:
        file.write(f"{student_id},{name}\n")
        
    print(f"✅ Student '{name}' registered successfully!")

def take_attendance():
    """Takes attendance for the day and saves it to the text file."""
    students = load_students()
    if not students:
        print("\n❌ No students registered yet. Please add students first.")
        return

    print("\n--- Take Attendance ---")
    today = datetime.now().strftime("%Y-%m-%d")
    print(f"Date: {today}\n")

    # 'a' mode to append today's records to the attendance file
    with open(ATTENDANCE_FILE, "a") as file:
        for student_id, name in students.items():
            while True:
                status = input(f"Is {name} (ID: {student_id}) Present? (P/A): ").strip().upper()
                if status in ['P', 'A']:
                    full_status = "Present" if status == 'P' else "Absent"
                    # Write the record as a comma-separated line
                    file.write(f"{today},{student_id},{name},{full_status}\n")
                    break
                else:
                    print("Invalid input. Please enter 'P' or 'A'.")
                    
    print(f"\n✅ Attendance for {today} has been saved successfully!")

def view_attendance():
    """Reads the text file and prints attendance in a neat format."""
    if not os.path.exists(ATTENDANCE_FILE):
        print("\n❌ No attendance records found.")
        return

    print("\n--- Attendance Records ---")
    print(f"{'Date':<15} | {'ID':<10} | {'Name':<20} | {'Status':<10}")
    print("-" * 62)
    
    with open(ATTENDANCE_FILE, "r") as file:
        for line in file:
            line = line.strip()
            if line:
                # Split the line into its 4 components
                date, student_id, name, status = line.split(",")
                print(f"{date:<15} | {student_id:<10} | {name:<20} | {status:<10}")

def main_menu():
    """Main program control loop."""
    while True:
        print("\n==============================")
        print("  STUDENT ATTENDANCE SYSTEM   ")
        print("==============================")
        print("1. Register a New Student")
        print("2. Take Attendance for Today")
        print("3. View Attendance History")
        print("4. Exit")
        
        choice = input("\nEnter your choice (1-4): ").strip()
        
        if choice == '1':
            add_student()
        elif choice == '2':
            take_attendance()
        elif choice == '3':
            view_attendance()
        elif choice == '4':
            print("\nThank you for using the Attendance System. Goodbye!")
            break
        else:
            print("❌ Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main_menu()
```
