```python
#count number of unique elements in a k window
from collections import Counter
def count_unique(nums,k):
    result=[]

    window_count=Counter(nums[:k])
    result.append(len(window_count))

    for i in range(k,len(nums)):
        element_leaving=nums[i-k]
        window_count[element_leaving]-=1

        if window_count[element_leaving]==0:
            del window_count[element_leaving]

        element_entering=nums[i]
        window_count[element_entering]+=1

        result.append(len(window_count))
    return result

nums=[1,2,1,3,4,2,3]
k=4
print(count_unique(nums,k))
```
