```python
#Minimum difference between highest and lowest score in k window
def min_diff(nums,k):
    if k==1:
        return 0
    nums.sort()

    min_diff=float('inf')

    for right in range(k-1,len(nums)):
        left=right-k+1
        current_diff=nums[right]-nums[left]
        min_diff=min(min_diff,current_diff)
    return min_diff

nums=[50,60,10,20,30,80]
k=3
print(min_diff(nums,k))
```
