```python
def decrypt(code,k):
    n=len(code)
    result=[0]*n
    if k==0:
        return result
    
    if k>0:
        current_window_sum=0
        for i in range(1,k+1):
            current_window_sum+=code[i%n]

        result[0]=current_window_sum

        for i in range(1,n):
            current_window_sum-=code[i]
            current_window_sum+=code[(i+k)%n]
            result[i]=current_window_sum
    return result
code=[5,7,1,4]
k=2
print(decrypt(code,k))
```
