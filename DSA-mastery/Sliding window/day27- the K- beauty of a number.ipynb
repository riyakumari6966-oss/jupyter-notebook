```python
#find the K beauty of a number
def k_beauty(num,k):
    str_num=str(num)
    count=0

    for i in range(len(str_num)-k+1):
        sub_str=str_num[i:i+k]
        sub_num=int(sub_str)

        if sub_num!=0 and num%sub_num==0:
            count+=1
    return count

num=430043
k=2
print(k_beauty(num,k))
```
