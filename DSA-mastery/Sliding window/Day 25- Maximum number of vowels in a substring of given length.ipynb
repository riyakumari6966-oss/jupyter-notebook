```python
#Maximum number of vowels in a substring of given length 
def max_vowels_inSubstring(s,k):
    list_s=list(s)
    vowels=set("aeiouAEIOU")
    count=0
    result=[]
    for i in range(k):
        if list_s[i] in vowels:
            count+=1
    result.append(count)

    for i in range(k,len(list_s)):
        char_leaving=list_s[i-k]
        if char_leaving in vowels:
            count-=1
        char_entering=list_s[i]
        if char_entering in vowels:
            count+=1
        result.append(count)
    return max(result)
s="Reshika"
k=3
max_vowels_inSubstring(s,k)
```
