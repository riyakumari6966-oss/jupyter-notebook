```python
#Maximum sum of ascending subarray
def max_ascending_sum(nums):
    current_sum=nums[0]
    max_sum=nums[0]
    for i in range(1,len(nums)):
        if nums[i]>nums[i-1]:
            current_sum+=nums[i]
        else:
            max_sum=max(current_sum,max_sum)
            current_sum=nums[i]
    return max(max_sum,current_sum)


nums=[10,20,30,4,40,50]
max_ascending_sum(nums)
```
