```python
#Count element less than X in every window
def count_element(arr,k,x):
    count=0
    result=[]
    for i in range(k):
        if arr[i]<x:
            count+=1
    result.append(count)

    for i in range(k,len(arr)):
        element_leaving=arr[i-k]
        if element_leaving<x:
            count-=1

        element_entering=arr[i]
        if element_entering<x:
            count+=1

        result.append(count)
    return result
            
            
         
arr=[1,2,3,5,6,8,0]
k=3
x=4
print(count_element(arr,k,x))
```
