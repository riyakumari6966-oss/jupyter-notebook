```python
#reverse word in a string as well as words
def reversestring(str):
    list_s=list(str)
    n=len(list_s)
    left=0
    for right in range(n+1):
        if right==n or right==' ':
            start=left
            end=right-1

            while start<end:
                list_s[start],list_s[end]=list_s[end],list_s[start]
                start+=1
                end-=1
            left=right+1
    return "".join(list_s)

str1="race is crazy"
reversestring(str1)
```

```python
#Reverse words in a string
def reverseWord(str):
    list_s=list(str)
    n=len(list_s)
    left=0
    for right in range(n+1):
        if right==n or list_s[right]==' ':
            start=left
            end=right-1

            while start<end:
                list_s[start],list_s[end]=list_s[end],list_s[start]
                start+=1
                end-=1

            left=right+1
    return "".join(list_s)

str1="race is crazy"
reverseWord(str1)
```
