```python
#find duplicate characters in a string
str1="Rishika"
str_list=list(str1.lower())
print(str_list)
count=0
for left in range(len(str_list)):
    for right in range(left+1,len(str_list)):
        if str_list[left]==str_list[right]:
            count+=1
            print([left,right]) 
            print("Duplicate character is:",str_list[left])
            print("Duplicate item is repeated:",count)
```

```python
#cleaner version
str1="Rishikaaaa"
str_list=list(str1.lower())
duplicates=[]
for left in range(len(str_list)):
    for right in range(left+1,len(str_list)):
        if str_list[left]==str_list[right]:
            if str_list[left] not in duplicates:
                duplicates.append(str_list[left])

print("Duplicates characters found:",duplicates)
```
