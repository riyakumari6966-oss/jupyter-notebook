```python
#find first palindromic string in an array
def firstpalindrome(words):
    def is_palindrome(word):
        left=0
        right=len(word)-1
        while left<right:
            if word[left]!=word[right]:
                return False
            left+=1
            right-=1
        return True
    for word in words:
        if is_palindrome(word):
            return word
    return ""

words=["abc","efg","car","racecar","efe","ghg"]
firstpalindrome(words)
```
