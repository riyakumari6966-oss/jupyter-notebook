```python
def are_rotations(s1,s2):
    n=len(s1)
    for i in range(n):
        if s1==s2:
            return True
        s1= s1[-1]+s1[:-1]  #s1[-1] return d + s1[:-1] means start from 0 index but it stops before index -1 return abc = dabc
        print(s1)
    return False

s1="abcd"
s2="cdab"
print(are_rotations(s1,s2))
```
