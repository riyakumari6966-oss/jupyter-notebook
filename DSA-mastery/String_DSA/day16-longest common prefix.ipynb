```python
def longestCommonPrefix(strs: list[str]) -> str:
    if not strs:
        return ""
    for col in range(len(strs[0])):
        print(col)
        char_to_check = strs[0][col]
        print(char_to_check)
        for row in range(1, len(strs)):
            if col == len(strs[row]) or strs[row][col] != char_to_check:
                return strs[0][:col]
    return strs[0]

strs1=["flower","floss","flow"]
print(longestCommonPrefix(strs1))
```
