```python
class ListNode:
    def __init__(self,x):
        self.val=x
        self.next=None
def Find_middle(head):
    slow=head
    fast=head

    while fast and fast.next:
        slow=slow.next
        fast=fast.next.next

    return slow
if __name__=='__main__':
    n1=ListNode(1)
    n2=ListNode(2)
    n3=ListNode(3)
    n4=ListNode(4)
    n5=ListNode(5)

    n1.next=n2
    n2.next=n3
    n3.next=n4
    n4.next=n5

    mid=Find_middle(n1)
    print(mid.val)
```
