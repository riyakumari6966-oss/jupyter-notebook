```python
#maximum count of positive or negative integers
def maxcount(nums):
    pos_no=0
    neg_no=0
    left=0
    while left<len(nums) and nums[left]<0:
        neg_no+=1
        left+=1
    right=len(nums)-1
    while right>=0 and nums[right]>0:
        pos_no+=1
        right-=1
    return max(pos_no,neg_no)
nums=[-3,-2,-1,0,1,2,3,4,5,6]
maxcount(nums)
```
