```python
# Merge two 2D arrays by summing values
def merge2Darray(nums1,nums2):
    n=len(nums1)
    m=len(nums2)
    p1=0
    p2=0
    result=[]
    while p1<n and p2<m:
        id1, val1=nums1[p1][0],nums1[p1][1]
        id2, val2=nums2[p2][0],nums2[p2][1]
        if id1==id2:
            result.append([id1,val1+val2])
            p1+=1
            p2+=1
        elif id1<id2:
            result.append([id1,val1])
            p1+=1
        else:
            result.append([id2,val2])
            p2+=1
    while p1<n:
        result.append(nums1[p1])
        p1+=1
    while p2<m:
        result.append(nums2[p2])
        p2+=1
    return result

nums1=[[1,2],[2,4],[4,2]]
nums2=[[1,3],[2,2],[3,6]]
merge2Darray(nums1,nums2)
```
