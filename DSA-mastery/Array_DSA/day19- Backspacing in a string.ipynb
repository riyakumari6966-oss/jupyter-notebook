```python
#Backspacing in a string
def backspacing(s,t):
    i=len(s)-1
    j=len(t)-1
    skip_s=0
    skip_t=0
    while i>=0 and j>=0:
        while i>=0:
            if s[i]=="#":
                skip_s+=1
                i-=1
            elif skip_s>0:
                skip_s-=1
                i-=1
            else:
                break
        while j>=0:
            if t[j]=="#":
                skip_t+=1
                j-=1
            elif skip_t>0:
                skip_t-=1
                j-=1
            else:
                break
        
        if i>=0 and j>=0:
            if s[i]!= t[j]:
                return(False)
        elif (i>=0) != (j>=0):
            return(False) 
        i-=1
        j-=1
    return (True)

s="ab#c"
t="ad#c"
backspacing(s,t)


            
```
