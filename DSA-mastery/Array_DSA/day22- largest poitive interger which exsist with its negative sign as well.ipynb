```python
#largest positive integer that exsist with its negative sign also
def findmaxk(nums):
    nums.sort()
    left=0
    right=len(nums)-1
    max_value=0
    while left<right:
        if abs(nums[left])>abs(nums[right]):
            max_value=abs(nums[left])
            left+=1
        elif abs(nums[left])<abs(nums[right]):
            max_value=abs(nums[right])
            right-=1
        else:
            if nums[left]==-nums[right]:
                return nums[right]
            left+=1
    return -1

nums=[1,-4,8,9,4,20]
findmaxk(nums)
```
