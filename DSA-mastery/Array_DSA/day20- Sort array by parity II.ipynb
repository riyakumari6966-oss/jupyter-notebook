```python
#sort array by parity II
def sortevenandodd(nums):
    n=len(nums)
    result=[0]*n
    even_idx=0
    odd_idx=1
    for num in nums:
        if num%2==0:
            result[even_idx]=num
            even_idx+=2
        else:
            result[odd_idx]=num
            odd_idx+=2
    return(result)

nums1=[2,5,7,8,10]
sortevenandodd(nums1)
```
