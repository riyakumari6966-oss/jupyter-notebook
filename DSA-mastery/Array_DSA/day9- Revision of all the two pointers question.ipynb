```python
#two sum problem
num=[2,7,11,15]
target=9
left=0
right=len(num)-1
while left<right:
    sum=num[left]+num[right]
    if sum==target:
        print([left,right])
        break
    elif sum<target:
        left+=1
    else:
        right-=1
    
```

```python
num=[3,2,4]
target=6
for i in range(len(num)):
    for j in range(i+1,len(num)):
        if num[i]+num[j]==target and i!=j:
            print([i,j])
            break
```

```python
#removing duplicates from sorted array
nums=[1,1,2,2,2,3,4,4,5]
left=0
for right in range(1,len(nums)):
    if nums[left]!=nums[right]:
        left+=1
        nums[left]=nums[right]
print(nums[:left+1])
```

```python
#moving zeroes to end of array
nums=[0,1,0,3,12]
left=0
for right in range(len(nums)):
    if nums[right]!=0:
        nums[left],nums[right]=nums[right],nums[left]
        left+=1
print(nums)
```

```python
#valid palindrome
s="race a car"
s="".join(s.lower() for char in s if char.isalnum())
left=0
right=len(s)-1
is_palindrome=True
while left<right:
    if s[left]!=s[right]:
        is_palindrome=False
        break
    left+=1
    right-=1
if is_palindrome:
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

```python
#squares of sorted array
nums=[-4,-1,0,3,10]
left=0
right=len(nums)-1
res=[0]*5
for i in range(len(res)-1,-1,-1):
    if abs(nums[left])>abs(nums[right]):
        res[i]=nums[left]**2
        left+=1
    else:
        res[i]=nums[right]**2
        right-=1
print(res)
```

```python
#container with most water
height=[1,8,6,2,5,4,8,3,7]
left=0
right=len(height)-1
max_area=0
while left<right:
    width=right-left
    height_of_container=min(height[left],height[right])
    area=width*height_of_container
    max_area=max(max_area,area)
    if height[left]<height[right]:
        left+=1
    else:
        right-=1

print(max_area)
```
