```python
#max sum of subarray of size k
def max_sum_subarray(arr,k):
    current_sum= sum(arr[:k])
    max_val=current_sum
    for i in range(k,len(arr)):
        current_sum+=arr[i]-arr[i-k]
        max_val=max(max_val,current_sum)
    return max_val

arr=[2,1,5,1,3,2]
k=3
print(max_sum_subarray(arr,k))
```

```python
def elements_max_sum_subarray(arr,k):
    current_sum= sum(arr[:k])
    max_val=current_sum
    best_start=0
    for i in range(k,len(arr)):
        current_sum+=arr[i]-arr[i-k]
        if current_sum>max_val:
            max_val=current_sum
            best_start=i-k+1
    return arr[best_start:best_start+k]

arr=[2,1,5,1,3,2]
k=3
print(elements_max_sum_subarray(arr,k))

```

```python
def max_product_subarray(arr,k):
    current_product=1
    for i in range(k):
        current_product*=arr[i]
    max_val=current_product
    for i in range(k,len(arr)):
        current_product*=arr[i]
        current_product//=arr[i-k]
        max_val=max(max_val,current_product)
    return max_val

arr=[2,1,5,1,3,2]
k=2
print(max_product_subarray(arr,k))
```
