```python
num=[-1,0,1,2,-1,-4]
num.sort()
print(num)
res=[]
for k in range(0,len(num)-1):
    if k>0 and num[k]==num[k-1]:
        continue
    i=k+1
    j=len(num)-1
    while i<j:
        if num[i]+num[k]+num[j]==0 and i!=j and j!=k and i!=k:
            res.append([num[k],num[i],num[j]])
            i+=1
            j-=1
            while i<j and num[i]==num[i-1]:
                i+=1
            while i<j and num[j]==num[j+1]:
                j-=1
        elif num[i]+num[j]+num[k]<0:
            i+=1
        else:
            j-=1
            print(res) 
```
