```python
num=[-1,0,1,2,-1,-4] 
num.sort() #sort the list to make it easier to find triplets
print(num) 
res=[]
for k in range(0,len(num)-1): #loop through the list and fix one element, then use two pointers to find the other two elements
    if k>0 and num[k]==num[k-1]:  #skip duplicates
        continue
    i=k+1               #left pointer
    j=len(num)-1        #right pointer
    while i<j:          #while the left pointer is less than the right pointer
        if num[i]+num[k]+num[j]==0 and i!=j and j!=k and i!=k: #if the sum of the three elements is zero and they are not the same element
            res.append([num[k],num[i],num[j]]) #add the triplet to the result list
            i+=1                      #move the left pointer to the right
            j-=1                      #move the right pointer to the left
            while i<j and num[i]==num[i-1]: #skip duplicates
                i+=1
            while i<j and num[j]==num[j+1]: #skip duplicates
                j-=1
        elif num[i]+num[j]+num[k]<0:
            i+=1
        else:
            j-=1
            print(res) 
```
