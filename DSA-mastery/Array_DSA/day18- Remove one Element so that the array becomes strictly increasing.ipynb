```python
#Remove one element such that array becomes strictly increasing
def Canbeincreasing(nums):
    count=0
    for i in range(1,len(nums)):
        if nums[i-1]>=nums[i]:
            count+=1
            
            if count>1:
                return(False)
                
                
            if i>1 and nums[i-2]>=nums[i]:
                nums[i]=nums[i-1]

    return(True)

nums=[2,3,1,2]
Canbeincreasing(nums)

```
