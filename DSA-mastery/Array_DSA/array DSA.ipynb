```python
# Move all zeros to the end of the array while maintaining the relative order of the non-zero elements.
nums=[0,1,0,2,12]
left=0                              # Initialize the left pointer to the first element of the array, which will be used to track the position of the last non-zero element.
for right in range(len(nums)):      # Start the right pointer from the first element and iterate through the array.
    if nums[right]!=0:              # If the current element is non-zero, we need to move it to the position of the last non-zero element.
        nums[left],nums[right]=nums[right],nums[left] # Swap the current non-zero element with the element at the left pointer, effectively moving the non-zero element to the front of the array.
        left+=1                    # Move the left pointer to the next position to prepare for the next non-zero element.
        print(nums)
print(nums)                        
```

```python
#Removing duplicates from a sorted array.
nums=[1,1,2,2,3,3,4,5]
left=0                           # Initialize the left pointer to the first element of the array, which will be used to track the position of the last unique element.
for right in range(1,len(nums)): # Start the right pointer from the second element and iterate through the array.
    if nums[right]!=nums[left]:  # If the current element is different from the last unique element, we have found a new unique element.
        left+=1                  # Move the left pointer to the next position to place the new unique element.
        nums[left]=nums[right]   # Update the value at the left pointer to the current unique element.
        print(nums)
print(nums[:left+1])     # The unique elements are now at the beginning of the array, and we return the length of the unique portion of the array.
```

```python
#also we can swap and remove duplicates
nums=[1,1,2,2,3,3,4,5]
left=0
for right in range(1,len(nums)):
    if nums[left]!=nums[right]:
        left+=1
        nums[left],nums[right]=nums[right],nums[left]
        print(nums)
print(nums[:left+1])
```

```python
#two sum problem (this method only works if the input array is sorted) O(n)
nums=[2,7,11,15]
target=13
left=0                                 # Initialize the left pointer to the first element of the array.
right=len(nums)-1                      # Initialize the right pointer to the last element of the array.
while left<right:                      # Continue the loop until the left pointer is less than the right pointer.
    current_sum=nums[left]+nums[right] # Calculate the sum of the elements at the left and right pointers.
    if current_sum==target:            # If the current sum is equal to the target, we have found the two numbers that add up to the target.
        print([left,right])            # Return the indices of the two numbers.
        break 
    elif current_sum>target:         
        right-=1                       
    else:
        left+=1                       

```

```python
#two sum problem (#for both sorted and unsorted array)
nums=[6,5,1,4,2,9,7]
target=3
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i]+nums[j]==target and i!=j:
            print([i,j])
```
