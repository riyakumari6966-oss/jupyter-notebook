```python
#correct approach
def max_area(height):
    left =0
    right = len(height) - 1
    max_area = 0

    while left < right:
        width=right-left
        height1=min(height[left],height[right])
        current_area=width*height1
        max_area=max(max_area,current_area)
        if height[left]<height[right]:
            left+=1

        else:
            right-=1

    return max_area

height2=[1,1,1,1,1,100,100,1,1]
print(max_area(height2))
```
