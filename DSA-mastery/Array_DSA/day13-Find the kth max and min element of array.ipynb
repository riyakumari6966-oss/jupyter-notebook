```python
#finding kth max and min element in an array
arr=[3,4,9,7,8,12,1]
k=2
arr.sort()
print(arr)
print("kth max element:",arr[-k])
print("kth min element:",arr[k-1])
    
```

```python
#using bubble sort to find kth max and min element in an array
arr=[3,4,9,7,8,12,1]
k=3
for i in range(len(arr)):
    for j in range(0,len(arr)-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]

kth_min=arr[k-1]
kth_max=arr[-k]

print(arr)
print(kth_min)
print(kth_max)
```
