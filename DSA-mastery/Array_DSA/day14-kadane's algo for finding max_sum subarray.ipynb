```python
#kadane's algo for finding max_sum_subarray
def max_subarray_sum(arr):
    current_max=arr[0]
    global_max=arr[0]

    for i in range(1,len(arr)):
        current_max=max(arr[i],current_max+arr[i])
        global_max=max(global_max,current_max)

    return global_max

arr1=[-2, 1, -3, 4, -1, 2, 1, -5, 4]
result=max_subarray_sum(arr1)
print(result)
```

```python
def max_subarray_with_indices(arr):
    current_max=arr[0]
    global_max=arr[0]
    start=end=temp_start=0

    for i in range(1,len(arr)):
        if arr[i]>current_max:
            current_max=arr[i]
            temp_start=i

        else:
            current_max+=arr[i]

        if current_max>global_max:
            global_max=current_max
            start=temp_start
            end=i

    return global_max,[start,end],arr[start:end+1]

arr1=[-2, 1, -3, 4, -1, 2, 1, -5, 4]
result=max_subarray_with_indices(arr1)
print(result)
```
