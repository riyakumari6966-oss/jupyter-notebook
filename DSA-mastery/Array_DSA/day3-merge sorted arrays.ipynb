```python
#correct approach
nums1=[1,2,3,0,0,0]  
m=3                #number of elements in nums1
nums2=[0,0,1]
n=3                #number of elements in nums2
i=m-1              #index of last element in nums1
j=n-1              #index of last element in nums2
k=m+n-1            #index of last position in nums1
while i>=0 and j>=0: #loop until we have elements in both arrays
    if nums1[i]>nums2[j]: #compare last elements of both arrays and place the larger one at the end of nums1
        nums1[k]=nums1[i] #place the larger element at the end of nums1
        i-=1              #move the index of nums1 to the left
    else:
        nums1[k]=nums2[j] #place the larger element at the end of nums1
        j-=1              #move the index of nums2 to the left
    k-=1                  #move the index of nums1 to the left
while j>=0:               #if there are still elements in nums2, place them in nums1
    nums1[k]=nums2[j]     #place the remaining element of nums2 in nums1
    j-=1                  #move the index of nums2 to the left
    k-=1                  #move the index of nums1 to the left
print(nums1)              #print the merged array
```
