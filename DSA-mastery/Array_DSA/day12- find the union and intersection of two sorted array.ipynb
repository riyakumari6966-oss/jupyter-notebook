```python
#finding union and intersection of two sorted arrays
num1=[1,2,4,5,6]
num2=[4,5,6,7,8]
union=[]
intersection=[]
i=0
j=0
while i<len(num1) and j<len(num2):
    if num1[i]<num2[j]:
        union.append(num1[i])
        i+=1
    elif num1[i]>num2[j]:
        union.append(num2[j])
        j+=1
    else:
        union.append(num1[i])
        intersection.append(num1[i])
        i+=1
        j+=1

while i<len(num1):
    union.append(num1[i])
    i+=1

while j<len(num2):
    union.append(num2[j])
    j+=1

print("Union of two sorted arrays is:",union)
print("Intersection of two sorted arrays is:",intersection)
```
