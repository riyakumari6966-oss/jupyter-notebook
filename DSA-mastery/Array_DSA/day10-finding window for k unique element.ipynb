```python
def longest_substring_k_distinct(s, k):
    left=0                 # Initialize the left pointer of the sliding window
    char_count={}          # Dictionary to count the occurrences of characters in the current window
    max_length=0           # Variable to keep track of the maximum length of a valid substring

    for right in range(len(s)):     # Iterate through the string with the right pointer
        char=s[right]               # Get the current character at the right pointer
        char_count[char]=char_count.get(char,0)+1   # Increment the count of the current character in the dictionary
        print(char_count)
        while len(char_count)>k:     # If the number of distinct characters exceeds k, shrink the window from the left
            left_char=s[left]        # Get the character at the left pointer
            char_count[left_char]-=1 # Decrement the count of the left character in the dictionary
            print(char_count)
            if char_count[left_char]==0:  # If the count of the left character becomes zero, remove it from the dictionary
                del char_count[left_char] # Remove the character from the dictionary
            left+=1                       # Move the left pointer to the right to shrink the window

        max_length=max(max_length,right-left+1) # Update the maximum length of a valid substring if the current window is larger
    return max_length

s=["A","A","B","C","B","B"]
k=2
result=longest_substring_k_distinct(s,k)
print(result)
```
