```python
#better approach
ph="race a car"
clean_ph="".join(char.lower() for char in ph if char.isalnum())
print(clean_ph)
left=0
right=len(clean_ph)-1
is_palindrome=True
while left<right:
    if clean_ph[left]!=clean_ph[right]:
        is_palindrome=False
        break
    left+=1
    right-=1

if is_palindrome:
    print("True,It is a palindrome")
else:
    print("False,Not a palindrome")
```
