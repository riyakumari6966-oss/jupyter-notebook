```python
#squares of sorted array
nums=[-8,-4,-2,3,5,6]
left=0
right=len(nums)-1
res=[0,0,0,0,0,0]
for p in range(len(res)-1,-1,-1):
    if abs(nums[left])>abs(nums[right]):
        res[p]=nums[left]**2
        left+=1
    else:
        res[p]=nums[right]**2
        right-=1
print(res)
```

```python
#valid palindrome
s="race a car"
s1=" ".join(char.lower() for char in s if char.isalnum())  #removing spaces and special characters and converting to lowercase
left=0
right=len(s1)-1
is_palindrome= True
while left<right:
    if s1[left]!=s1[right]:
        is_palindrome=False
        break
    left+=1
    right-=1

if is_palindrome:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
    
        
```
